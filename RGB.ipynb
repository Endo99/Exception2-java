```
import numpy as np
from PIL import Image

from matplotlib import image
from matplotlib import pyplot as plt
```

```
img_s = image.imread('/content/sebastian-unrau-v4e3JI7DDHI-unsplash.jpg')
img_s
```

```
img_origin = Image.open('/content/sebastian-unrau-v4e3JI7DDHI-unsplash.jpg')
img_origin
```

```
data = np.array(img_origin)
```

```
plt.figure(figsize=(16, 9))
plt.imshow(img_origin)
plt.axis('off')
plt.show()
```

```
new_img = data.copy()
```

```
print(img_origin.size)
```

```
img_red = new_img[:,:,0]
img_green = new_img[:,:,1]
img_blue = new_img[:,:,2]
```

```
fig, (ax_red, ax_green, ax_blue) = plt.subplots(1, 3, figsize=(16, 9))

ax_red.imshow(img_red)
ax_red.axis('off')
ax_green.imshow(img_green)
ax_green.axis('off')
ax_blue.imshow(img_blue)
ax_blue.axis('off')

plt.show()
```
